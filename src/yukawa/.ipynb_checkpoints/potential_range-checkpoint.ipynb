```python
path = '../../data/yukawa/k=10.0.txt'

# initial config
r_init = np.linspace(0.1, 8, 200)
step = r_init[-2] - r_init[-3]

df = pd.read_csv(path, sep="\s+", names=["k", "q", "delta_0", "delta_1", "delta_2"])

results = {}

for r_max in [2.0, 4.0, 6.0, 8.0]:
    print('Computing', r_max)
    print()

    r = np.arange(0.1, r_max + step, step) # 200-grain 

    V_vec = []
    for i in range(len(df)):
        q = df.iloc[i]['q']
        v = np.array([yukawa(q, ri, V_0=5) for ri in r])
        V_vec.append(v)

    df['V'] = V_vec

    X = df['V'].values
    y = df['delta_0'].values

    X = np.array([np.array(x) for x in X])
    X = np.expand_dims(X, axis=-1)
    y = np.array(y)

    # train / test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
    print(X_train.shape)
    
    # load the model
    model, model_name = cnn(input_shape=(X.shape[1], 1))

    # fit model
    history = model.fit(
        X_train,
        y_train,
        batch_size=16,
        epochs=100,
        validation_split=0.3,
        verbose=0
        )

    results[r_max_str] = {}
    results[r_max_str]['r'] = r
    results[r_max_str]['V'] = V_vec
    results[r_max_str]['predictions'] = np.array([p[0] for p in model.predict(X_test)]).flatten()
```

```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 10))

for i, k in enumerate(results):    
    r = results[k]['r']
    V = results[k]['V'][0]
    predictions = results[k]['predictions']
    
    ax1.plot(r, V, label='$r_{max}0=$%s'%k)
    ax2.scatter(y_test, predictions, label='$r_{max}0=$%s'%k, s=1.)

ax2.plot([0, 1], [0, 1], transform=ax2.transAxes, linestyle='dashed', color='k')

ax1.legend()
ax2.legend()

plt.show()
```
